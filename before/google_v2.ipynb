## 가장 많이 반복된 문장 파악

데이터셋을 생성하기 위해 한글, 영어 이중언어를 하나로 통합시킬 필요가 있음.  
많이 반복된 문장, 단어를 파악하여 효율적으로 번역하는 것을 목표로 함.  
단어 변경을 진행하기 앞서 중복되는 문장도 많은 것으로 확인되어 문장 단위의 변경을 우선.

```python
import re
```

```python
finding = dataset['Findings'].str.replace('\r','\n').values.tolist()
sentence_list = []
for information in finding:
    try:
        # 줄바꿈 기준으로 분리
        sentence_list =sentence_list + information.split('\n')
    except:
        pass

# 한글이 포함된 문장의 등장 횟수 체크
kor_count_dict = dict()
for x in set(sentence_list):
    if re.search('[가-힣]',x) == None:
        continue
    kor_count_dict[x]=sentence_list.count(x)
sorted(kor_count_dict.items(), key = lambda item: item[1], reverse = True)
```

### 영상 촬영에 대한 내용이 많음. 동일하게 처리할 수 있을 것 같아 빼고 따로 볼 예정

```python
kor_count_dict2 = dict()
for x in set(sentence_list):
    if re.search('[가-힣]',x) == None:
        continue
    elif '조영증강' in x:
        continue
    kor_count_dict2[x]=sentence_list.count(x)
sorted(kor_count_dict2.items(), key = lambda item: item[1], reverse = True)
```

>문장 단위로 교체를 하려 하였으나, 영상 촬영 말고는 반복되지 않는 문장이나 단어가 많았다.  
그래서 구글 번역기를 이용하여 번역을 진행해볼까 한다. 영어 + 한글 -> 영어로 진행할 것이다.  
판독문에 기술된 혼용언어는 영어 단어를 사용한 국어 문장인 경우가 대다수였다.  
몇 문장을 시행해본 결과 번역이 잘 되는 것을 확인하여 전체에 진행해보게 되었다.
